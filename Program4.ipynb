```python
# Program 4
# CIS 443-01
# Due by: 4/25/2022
# Written by: C7933

""" This program will analyze Disneyland Trip Advisor reviews comparing the sentiments of star ratings and text reviews"""

# import libraries
import pandas as pd
from textblob import TextBlob
import numpy as np

# load csv into a pandas dataframe
df_Disneyland = pd.read_csv('DisneylandReviews.csv', encoding='latin-1')

# create function to classify star rating sentiment
def add_rating_sentiment(rating):
    if rating == 4 or rating == 5:
        return 'Positive'
    elif rating == 3:
        return 'Neutral'
    else:
        return'Negative'

# add column for star rating sentiment classification
df_Disneyland['Star_Sentiment'] = [add_rating_sentiment(rating) for rating in df_Disneyland['Rating']]

# calculate each review's sentiment polarity using each review's text and add this result to a new column
df_Disneyland['Review_Polarity'] = [TextBlob(review).sentiment.polarity for review in df_Disneyland['Review_Text']]

# create function to classify text review sentiment
def add_review_sentiment(polarity):
    if polarity > 0.05:
        return 'Positive'
    elif polarity < -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# add column for text review sentiment classification
df_Disneyland['Review_Sentiment'] = [add_review_sentiment(polarity) for polarity in df_Disneyland['Review_Polarity']]


# determine whether star rating sentiment classification agrees with the text review sentiment classification and add this result to a new column
df_Disneyland['Sentiment_Agreement'] = [agreement for agreement in np.where(df_Disneyland['Star_Sentiment'] == df_Disneyland['Review_Sentiment'], True, False)]
print(df_Disneyland)

# output a summary table 1
true_df_Disneyland = df_Disneyland[df_Disneyland.Sentiment_Agreement == True]
false_df_Disneyland = df_Disneyland[df_Disneyland.Sentiment_Agreement == False]

print(f'\nNumber agree:    {len(true_df_Disneyland)} {len(true_df_Disneyland)/len(df_Disneyland):.2%}')
print(f'Number disagree: {len(false_df_Disneyland)} {len(false_df_Disneyland)/len(df_Disneyland):.2%}\n')

# output a summary table 2
print('Review Stars|Total Reviews|Num Agree|% Agree|Num Disagree|% Disagree')
print('------------|-------------|---------|-------|------------|----------')
print(f'{5:^12}|{len(df_Disneyland.query("Rating==5")):^13}|{len(df_Disneyland.query("Rating==5 & Sentiment_Agreement==True")):^9}|{len(df_Disneyland.query("Rating==5 & Sentiment_Agreement==True"))/len(df_Disneyland.query("Rating==5")):7.2%}|{len(df_Disneyland.query("Rating==5 & Sentiment_Agreement==False")):^12}|{len(df_Disneyland.query("Rating==5 & Sentiment_Agreement==False"))/len(df_Disneyland.query("Rating==5")):7.2%}')
print(f'{4:^12}|{len(df_Disneyland.query("Rating==4")):^13}|{len(df_Disneyland.query("Rating==4 & Sentiment_Agreement==True")):^9}|{len(df_Disneyland.query("Rating==4 & Sentiment_Agreement==True"))/len(df_Disneyland.query("Rating==4")):7.2%}|{len(df_Disneyland.query("Rating==4 & Sentiment_Agreement==False")):^12}|{len(df_Disneyland.query("Rating==4 & Sentiment_Agreement==False"))/len(df_Disneyland.query("Rating==4")):7.2%}')
print(f'{3:^12}|{len(df_Disneyland.query("Rating==3")):^13}|{len(df_Disneyland.query("Rating==3 & Sentiment_Agreement==True")):^9}|{len(df_Disneyland.query("Rating==3 & Sentiment_Agreement==True"))/len(df_Disneyland.query("Rating==3")):7.2%}|{len(df_Disneyland.query("Rating==3 & Sentiment_Agreement==False")):^12}|{len(df_Disneyland.query("Rating==3 & Sentiment_Agreement==False"))/len(df_Disneyland.query("Rating==3")):7.2%}')
print(f'{2:^12}|{len(df_Disneyland.query("Rating==2")):^13}|{len(df_Disneyland.query("Rating==2 & Sentiment_Agreement==True")):^9}|{len(df_Disneyland.query("Rating==2 & Sentiment_Agreement==True"))/len(df_Disneyland.query("Rating==2")):7.2%}|{len(df_Disneyland.query("Rating==2 & Sentiment_Agreement==False")):^12}|{len(df_Disneyland.query("Rating==2 & Sentiment_Agreement==False"))/len(df_Disneyland.query("Rating==2")):7.2%}')
print(f'{1:^12}|{len(df_Disneyland.query("Rating==1")):^13}|{len(df_Disneyland.query("Rating==1 & Sentiment_Agreement==True")):^9}|{len(df_Disneyland.query("Rating==1 & Sentiment_Agreement==True"))/len(df_Disneyland.query("Rating==1")):7.2%}|{len(df_Disneyland.query("Rating==1 & Sentiment_Agreement==False")):^12}|{len(df_Disneyland.query("Rating==1 & Sentiment_Agreement==False"))/len(df_Disneyland.query("Rating==1")):7.2%}')
```
